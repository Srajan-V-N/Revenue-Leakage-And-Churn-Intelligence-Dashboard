```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("../data/processed/telco_churn_cleaned.csv")
df.info()
```

```python
engine = create_engine(
    "mysql+pymysql://root:root@localhost:3306/revenue_leakage",
    echo=False
)
```

```python
df.to_sql(
    name="telco_churn",
    con=engine,
    if_exists="replace",   # ðŸ‘ˆ THIS FIXES EVERYTHING
    index=False,
    chunksize=1000         # ðŸ‘ˆ avoids packet-size errors
)
```

```python
pd.read_sql("SELECT COUNT(*) FROM telco_churn", engine)
```
